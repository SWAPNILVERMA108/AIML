```python
import json
```

```python
# load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data

load_data("store_data.json")

```

```python
#or store data in another variable
data1=load_data("store_data.json")
print(data1)
print(type(data1))
```

```python
# clean the data
def clean_data(data1):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    for user in data1:
        #clean rating
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"]=raw_rating
        # handling missing value
        raw_age = user.get("age")
        if(raw_age ==None):
            user["age"]=None
        #Deduplication
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])

        cleaned_data.append(user)
    return cleaned_data
            
        
            
        
```

```python
clean_data(data1)
```

```python
# Get meaningful insights from data
def get_insights(data1):

    # avg rating
    tot_rating = 0

    for user in data1:
        tot_rating += float(user["rating"])

    print(f"avg rating = {tot_rating/len(data1)}")


    # percentage of user with poor rating
    poor_ratings = 0

    for user in data1:
        if float(user["rating"]) < 3:
            poor_ratings += 1


    
    print(f"% of user with poor rating : {(poor_ratings/len(data1))*100}")
```

```python
get_insights(data1)
```

```python
# Build recommendation feature
def get_recommendations(data1):
    recommendations = [] 
    for user in data1:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "apple"
        else:
            curr_recomm["brand"] = "samsung"
        recommendations.append(curr_recomm) 
    return recommendations
        
    
    
```

```python
get_recommendations(data1)
```
